```python
#Câu 14:
def Luminosity(img):
  img = img.reshape(1, 1, -1)
  return img[:, :, 0] * 0.21 + img[:, :, 1] * 0.72 + img[:, :, 2] * 0.07

gray_img_03 = np.apply_along_axis(Luminosity, 2, img)
gray_img_03[0, 0]
```

```python
#Câu 15:
import pandas as pd
df = pd.read_csv('/content/advertising.csv')
data = df.to_numpy()

for i in range(len(data)):
  if data[i, 3] == np.max(data[:, 3]):
    print(f'Sales: {data[i, 3]} and Index:{i}')

```

```python
#Câu 16:
tv = np.average(data[:, 0])
print(tv)
```

```python
#Câu 17:
count = 0
for sales in data[:, 3]:
  if sales >= 20:
    count+=1
print(count)
```

```python
#Câu 18:
sales_column = data[:, 3]
radio_column = data[:, 1]
filtered_radio = radio_column[sales_column >= 15]
average_radio = np.mean(filtered_radio)
print(average_radio)
```

```python
#Câu 19:
newspaper_column = data[:, 2]
average_newspaper = np.mean(newspaper_column)
filtered_sales = sales_column[newspaper_column > average_newspaper]
total_sales = np.sum(filtered_sales)
print(total_sales)
```

```python
#Câu 20:
average_A = np.mean(data[:, 3])
scores = []
for sales in data[:, 3]:
  if sales > average_A:
    scores.append('Good')
  elif sales < average_A:
    scores.append('Bad')
  else:
    scores.append('Average')
print(scores[7:10])
```
